```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Bar Chart For top Sector by Total FDI
plt.figure(figsize=(14,8))
top_sectors = sorted_df.head(10)
sns.barplot(x=top_sectors['Total FDI'], y=top_sectors['Secotor'], palette='viridis')
plt.title('Top 10 Sectors by Total FDI (2000-01 to 2016-17)')
plt.xlabel('Total FDI (in million USD)')
plt.ylabel('Sector')
plt.show() 
```

```python
# Line Chart for Yearly FDI Trends
plt.figure(figsize=(14, 8))
yearly_fdi.plot(kind='line', marker='o', color='b')
plt.title('Yearly FDI Trends (2000-01 to 2016-17)')
plt.xlabel('Year')
plt.ylabel('FDI (in million USD)')
plt.grid(True)
plt.show()
```

```python
# Stacked Bar Chart for Sector-wise Yearly FDI
plt.figure(figsize=(14, 8))
df.set_index('Secotor').T.iloc[1:-2].plot(kind='bar', stacked=True, figsize=(14, 8), colormap='tab20')
plt.title('Sector-wise Yearly FDI (2000-01 to 2016-17)')
plt.xlabel('Year')
plt.ylabel('FDI (in million USD)')
plt.legend(title='Sector', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
